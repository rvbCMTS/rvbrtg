```python
# Match and identify nr of duplicate acq.prot for each studyUID, 
duplicates = ft_merged.groupby('studyInstanceUID')['acquisitionProtocol'].apply(lambda x: x.duplicated().sum())

# Identify which UIDs have duplicates, and print
# Next is to save it rather than only print it to terminal
duplicate_ids = duplicates[duplicates > 0]
if not duplicate_ids.empty:
    print("Duplicate datapoints found for the following IDs:")
    for id, count in duplicate_ids.items():
        print(f"ID {id} has {count} duplicate datapoints")
else:
    print("  No duplicates found for the selected IDs.")
```

```python
# Estimate ratio of examinations that include one or more re-takes,

# nr of examinations including re-takes divided by total nr of examinations
P = len(duplicate_ids)/len(duplicates)
print(f"About {100*P} % of all examinations have some >1 amount of retake images")
```

Code above provides an estimation on the total percentage of examinations that contain some-sort of "retake" image. 
A "retake" is defined in code as a duplicate acq.prot matching one studyUID. 
    - Since this is based on the total amount of examinations it provides a misleading result. Its more of interest so see the percentage based on individual exposures.

Next step is to estimate the ratio of exposures that are retakes. 

```python
# Estimate ratio of extra exposures that are due to re-takes,

# Find total amount of extra exposures
retakes = duplicate_ids.sum()

# nr of retake exposures divided by total nr of exposures
P2 = retakes/len(ft_merged)
print(f"About {100*P2} % of all exposures are retakes")
```

```python
# Estimate amount of duplicates for each type of acq.prot
duplicates_temp = ft_merged.groupby('acquisitionProtocol')['studyInstanceUID'].apply(lambda x: x.duplicated().sum())

# Estimate tot amount of each duplicate acq.prot
acqprot_counts = ft_merged['acquisitionProtocol'].value_counts()

# Calc percentage of duplicates for each datatype
temp_temp = 100*(duplicates_temp / acqprot_counts)
```

```python
# Pivot test
pivot = ft_merged.pivot_table(
    index='acquisitionProtocol',
    values='studyInstanceUID',
    aggfunc=[
        lambda x: x.duplicated().sum(),
        'count'
    ]
)
# Rename columns
pivot.columns = ['duplicates', 'total_count']
# Calc percentage
pivot['duplicate_percentage'] = 100 * (pivot['duplicates'] / pivot['total_count'])

# Print pivot table
print(pivot)
```
