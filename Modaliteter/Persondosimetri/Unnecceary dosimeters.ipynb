Exporterat data från SSP som csv. Vissa år var korrupta och kunde inte exporteras. Läser in data till pandas

```python
import os
import pandas as pd

csv_path = "C:\slask\persondoser"

data = pd.DataFrame()
for filename in os.listdir(csv_path):
    file_data = pd.read_csv(f'{csv_path}/{filename}')
    data = pd.concat([data, file_data], ignore_index=True)

```

Väljer enbart helkropps-dosimeter då ringar och ögondosimeter alltid är motiverade

```python
filtered_data = data[data['dosimeter_placement'] =='Helkropp']
filtered_data
```

Pivot-tabell för att sammanställa data

```python
summery_data = filtered_data.groupby(['personnel_identifier']).agg(
    total_hp10 = ("hp10", "sum"),
    total_hp007 = ("hp007", "sum"),
    number_of_months = ("hp10", "count"),
    latest_reading = ("center_time", "max")
)
```
